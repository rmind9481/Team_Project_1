```python
# 2023년도 뽑아내기
Traiifc_DF.columns
Traiifc_DF23 = Traiifc_DF[Traiifc_DF['년'] == 2023]

Traiifc_DF23
```

```python
# 2. '년'과 '월' 컬럼 제외 후 합계 계산

column_sums = Traiifc_DF23.drop(columns=['년', '월']).sum()
```

```python

# 3. 최대값과 최소값을 가진 컬럼 찾기
max_col = column_sums.idxmax()  # 가장 합계가 큰 컬럼
max_value = column_sums.max()   # 가장 큰 값

min_col = column_sums.idxmin()  # 가장 합계가 작은 컬럼
min_value = column_sums.min()   # 가장 작은 값  
```

```python
# 4. 결과 출력
print(f"가장 값이 많은 컬럼: {max_col}, 총합: {max_value}")
print(f"가장 값이 적은 컬럼: {min_col}, 총합: {min_value}")
percent = (min_value/max_value)*100
print(f'반월당 대비 학정의 이동평균 {percent: .3f}%')
```
