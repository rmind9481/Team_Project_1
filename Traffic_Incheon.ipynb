```python

df3 = Traiifc_DF[['정류소명', '총승차건수', '총하차건수']]
```

```python
df3['총승하차'] = df3['총승차건수']+df3['총하차건수']
# df3.set_index('정류소명',inplace=True)
df3
```

```python
# 중복되는 정류소명 그룹화 해서 총하자 건수 또한 합침
grouped = df3.groupby('정류소명', as_index=False)['총승하차'].sum()

print(type(grouped))


# 대구와 조건을 비슷하게 하기위해 하위 유동인구 10만개 정도로 추출
grouped = grouped[grouped['총승하차'].dropna() > 100000]

```

```python
# max_station = grouped.loc[grouped['총승하차'].idxmax()]
# min_station = grouped.loc[grouped['총승하차'].idxmin(skipna =True)]

max_station = grouped.loc[grouped['총승하차'].idxmax(skipna = True)]
# min_station = grouped.loc[grouped['총승하차'].idxmin(skipna =True)]

min_station = grouped.nsmallest(10,'총승하차')

print("일평균 승하차 건수가 가장 많은 정류소:")
print(max_station)

print("\n일평균 승하차 건수가 가장 적은 정류소:")
print(min_station)
```
